```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_mcdonalds = pd.read_csv('menu.csv')
```

```python
print(data_mcdonalds.head())
```

```python
print(data_mcdonalds.isnull().sum())
```

```python
stats = data_mcdonalds.describe()
print(stats)
```

```python
categ_analysis = data_mcdonalds.groupby('Category').agg({'Calories': ['mean', 'median'], 'Total Fat': 'mean', 'Sodium': 'mean'})
print(categ_analysis)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=categ_analysis.index, y=categ_analysis[('Calories', 'mean')])
plt.title('Avg Calories by Category')
plt.xlabel('Category')
plt.ylabel('Avg Calories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
numeric_columns = data_mcdonalds.select_dtypes(include=np.number)
plt.figure(figsize=(10, 8))
sns.heatmap(numeric_columns.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
